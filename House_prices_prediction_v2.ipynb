```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
def preprocessing(filename):
  data = pd.read_csv(filename)
  #lay ra du lieu y_train
  y = data['SalePrice'].values.reshape(-1,1)
  del data['SalePrice']
  # dien du lieu lien tuc bi miss
  for col in data.columns:
    if data[col][0].__class__ ==np.int64 or data[col][0].__class__ ==np.float64:
      data[col].fillna(data[col].mean(),inplace=True)
  # dien du lieu roi rac bi miss (du lieu dang str, nan la float)
  for col in data.columns:
    if data[col][0].__class__ == float or data[col][0].__class__ == str:
      data[col].fillna(data[col].mode()[0], inplace=True)
  # onehot du lieu roi rac
  for col in data.columns:
    if data[col][0].__class__ ==str or data[col][0].__class__ ==float:
      enc = OneHotEncoder()
      enc_df = pd.DataFrame(enc.fit_transform(data[[col]]).toarray(),
                        columns=np.sort(data[col].unique())+col)
      data= data.join(enc_df)
      del data[col]
  #du lieu train
  values = data.values
  N = values.shape[0]
  M = values.shape[1]+1
  x = values

  return N,M,x,y
```

```python
N,M,X,y = preprocessing('/content/drive/MyDrive/VCCorp/house-prices-prediction/train.csv')
```

```python
# tach tap train va test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 150)

print('Train size: ',len(y_train))
print('Test size: ',len(y_test))
```

```python
# Linear Regression voi thu vien sklearn
from sklearn.linear_model import LinearRegression
```

```python
# tao mo hinh hoi quy tuyen tinh
lrg = LinearRegression()
#train mo hinh voi data dat
lrg.fit(X_train, y_train)
```

```python
#du doan
y_pred = lrg.predict(X_train)
y_pred_test = lrg.predict(X_test)
```

```python
loss = 0

for i in range(len(y_train)):
  loss += abs(y_train[i][0]-y_pred[i][0])/y_train[i][0]

loss = loss/len(y_train)
accuracy = 100-loss*100

print('Do chuan xac cua mo hinh voi tap train la accuracy',accuracy)
```
